# Vector Plotting


```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
plt.quiver(0,0,3,4)
plt.show()
```

```
plt.quiver(0,0,3,4 , scale_units ='xy' ,angles = 'xy',scale =1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
plt.quiver(0,0,3,4 , scale_units ='xy' ,angles = 'xy',scale =1 , color = 'r')
plt.quiver(0,0,-3,4 , scale_units ='xy' ,angles = 'xy',scale =1, color = 'g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
def plot_vectors(vecs):
  colors = ['r','b','g','y']
  i=0
  for vec in vecs:
    plt.quiver(vec[0],vec[1],vec[2],vec[3], scale_units ='xy' ,angles = 'xy',scale =1, color = colors[i%len(colors)])
    i=i+1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
plot_vectors([(0,0,3,4),(0,0,-3,4),(0,0,4,3),(0,0,2,4)])
```

```
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,4,3]),np.asarray([0,0,2,4])]
```

```
plot_vectors([vecs[0],vecs[3]])
```

#Vector Addition

```
vecs[0]+vecs[1]
```

```
plot_vectors([vecs[0],vecs[3],vecs[0]+vecs[3]])
```

#Vector dot Product

```
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,4,3]),np.asarray([0,0,2,4])]
```

```
plot_vectors(vecs)
```

```
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

```
a_dot_b = np.dot(a,b)
```

```
print(a_dot_b)
```

$\vec{a}\cdot\vec{b} = |\vec{a}||\vec{b}|\cos(\theta) $

```
a_b = np.dot(a,b)/np.linalg.norm(b)
```

```
print(a_b)
```

```
vec_a_b = (a_b/np.linalg.norm(b))*b
```

```
print(vec_a_b)
```
